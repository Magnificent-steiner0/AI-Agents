```python
from langchain.chat_models import ChatOpenAI
from langchain.agents import initialize_agent, Tool, AgentType
from langchain.memory import ConversationBufferMemory
import os
```

```python
from google.colab import userdata
os.environ["OPENAI_API_KEY"] = userdata.get('writing-agent-key')
```

```python
def grammar_corrector(text: str)->str:
  return llm.predict(f"Correct the grammar in the following text: \n{text}")

def formal_rewriter(text: str)->str:
  return llm.predict(f"Rewrite the following text into formal academic text: \n{text}")

def simplify_text(text: str)->str:
  return llm.predict(f"Rewrite the following text to be simpler and readable for general audience: \n{text}")

def complete_article(text: str)->str:
  prompt=f"""
  You are a professional content writer. Complete the article based on following introduction.

  Introduction:{text}

  Continue writing:
  """
  return llm.predict(prompt.strip())
```

```python
tools=[
    Tool(
        name="GrammarFixer",
        func=grammar_corrector,
        description="Use this to fix grammar and punctuation mistakes"
    ),
        Tool(
        name="FormalRewriter",
        func=formal_rewriter,
        description="Use this to rewrite given text into a formal academic text"
    ),
        Tool(
        name="TextSimplifier",
        func=simplify_text,
        description="Use this to simplify the text and easier to read"
    ),
        Tool(
        name="ArticleFinisher",
        func=complete_article,
        description="Use this to finish the article given the intro"
    ),

]
```

```python
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.5)

agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent= AgentType.OPENAI_FUNCTIONS,
    verbose=True
)
```

```python
query = """ 'the system have many issue like sometimes it crash when user try to login, and also not good user experience'
can you correct the grammar?
"""
response = agent.run(f"As a grammar corrector, answer: {query}")


query = """ 'Rewrite this in formal tone: I wanna let you know that the project ain't gonna be done on time.'
"""
response = agent.run(f"As a format text writer, answer: {query}")


query = """ 'Simplify this: The proliferation of digital misinformation has exacerbated public distrust in traditional institutions.'
"""
response = agent.run(f"As a text simplifier, answer: {query}")


query = """ 'Complete this article: Remote work has become increasingly popular over the last few years, especially after the COVID-19 pandemic.'
"""
response = agent.run(f"As an article finisher, answer: {query}")
```
